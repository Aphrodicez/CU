```python
import http.client
def GET():
    # Connect to twitter.com over HTTPS
    conn = http.client.HTTPSConnection("twitter.com", 443)

    # Send a HTTP/1.0 request to "/"
    conn.putrequest("GET", "/", skip_host=True, skip_accept_encoding=True)

    # Add minimal headers (HTTP/1.0 does not require Host, but some servers might)
    conn.putheader("Host", "twitter.com")
    conn.endheaders()

    # Get response
    response = conn.getresponse()
    print("Status:", response.status, response.reason)
    print("Headers:")
    print(response.headers)
    print("\nBody:")
    print(response.read().decode(errors="replace"))
```
